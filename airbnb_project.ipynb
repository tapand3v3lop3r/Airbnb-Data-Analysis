# Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading the files

```python
df = pd.read_csv("Airbnb_csv.csv", low_memory = False)
```

# Handling Missing Values

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month' : 0, 'last review' : df['last review'].min()}, inplace = True)
```

```python
df.dropna(subset = ['NAME', 'host name'], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns = ['license', 'house_rules'], errors = 'ignore')
```

```python
# remove currency signs and convert to float

df['price'] = df['price'].replace('[$,]', '', regex = True).astype(float) 
df['service fee'] = df['service fee'].replace('[$,]', '', regex = True).astype(float) 
```

# Removal of Duplicates

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

# Data Visualisation

### What is the distribution of listing prices?

```python
plt.figure(figsize = (12, 8))
sns.histplot(df['price'], bins = 50, kde = True)
plt.title("Distribution of Listing Prices")
plt.xlabel("Price $")
plt.ylabel("Frequency")
plt.show()
```

### How are diifferent room types distributed?

```python
df['room type']
```

```python
plt.figure(figsize = (8, 5))
sns.countplot(x = 'room type', data = df, color = 'hotpink')
plt.title("Room Type Distributions")
plt.xlabel("Room Types")
plt.ylabel("Counts")
plt.show()
```

### How are listings distributed across different neighbourhoods?

```python
plt.figure(figsize = (12, 8))
sns.countplot(y = 'neighbourhood group', data = df, color = 'lightgreen', order = df['neighbourhood group'].value_counts().index)
plt.title("Number of Listings by different Neighborhood Groups")
plt.xlabel("Count")
plt.ylabel("Neighbourhood Groups")
plt.show()
```

### What is the relationship between Price and Room Type?

```python
plt.figure(figsize = (10, 8))
sns.boxplot(x = 'room type', y = 'price', hue = 'room type', data = df, palette = 'Set1')
plt.title("Price vs Room Type")
plt.xlabel("Room Type")
plt.ylabel("Price ($)")
plt.show()
```

### How has the number of reviews changed over with time?

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize = (12, 6))
reviews_over_time.plot(kind = 'line', color = 'red')
plt.title('Number of reviews over time')
plt.xlabel("Years")
plt.ylabel("Number of reviews")
plt.show()
```
